```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['STSong']
plt.rcParams['axes.unicode_minus'] = False
```

```python

# 读取Excel文件
excel_file = 'trainset.xlsx'

# 读取三个表格
df1 = pd.read_excel(excel_file, sheet_name='材料1')
df2 = pd.read_excel(excel_file, sheet_name='材料2')
df3 = pd.read_excel(excel_file, sheet_name='材料3')
df4 = pd.read_excel(excel_file, sheet_name='材料4')

# 合并表格
df = pd.concat([df1, df2, df3, df4])
```

```python
# 同一波形下，不同材料和温度对磁性损耗的影响
bx = '梯形波'
# 材料1
X1 = df1[['温度，oC','励磁波形','磁芯损耗，w/m3']]
X11 = X1[X1['励磁波形']==bx]

X1_tem25 = X11[X11['温度，oC']==25]
X1_tem50 = X11[X11['温度，oC']==50]
X1_tem70 = X11[X11['温度，oC']==70]
X1_tem90 = X11[X11['温度，oC']==90]

P1_tem25 = X1_tem25['磁芯损耗，w/m3']
P1_tem50 = X1_tem50['磁芯损耗，w/m3']
P1_tem70 = X1_tem70['磁芯损耗，w/m3']
P1_tem90 = X1_tem90['磁芯损耗，w/m3']

P1_tem25.columns = ['25度']
P1_tem50.columns = ['50度']
P1_tem70.columns = ['70度']
P1_tem90.columns = ['90度']
# 材料2
X2 = df2[['温度，oC','励磁波形','磁芯损耗，w/m3']]
X22 = X2[X2['励磁波形']==bx]

X2_tem25 = X22[X22['温度，oC']==25]
X2_tem50 = X22[X22['温度，oC']==50]
X2_tem70 = X22[X22['温度，oC']==70]
X2_tem90 = X22[X22['温度，oC']==90]

P2_tem25 = X2_tem25['磁芯损耗，w/m3']
P2_tem50 = X2_tem50['磁芯损耗，w/m3']
P2_tem70 = X2_tem70['磁芯损耗，w/m3']
P2_tem90 = X2_tem90['磁芯损耗，w/m3']

P2_tem25.columns = ['25度']
P2_tem50.columns = ['50度']
P2_tem70.columns = ['70度']
P2_tem90.columns = ['90度']
# 材料3
X3 = df3[['温度，oC','励磁波形','磁芯损耗，w/m3']]
X33 = X3[X3['励磁波形']==bx]

X3_tem25 = X33[X33['温度，oC']==25]
X3_tem50 = X33[X33['温度，oC']==50]
X3_tem70 = X33[X33['温度，oC']==70]
X3_tem90 = X33[X33['温度，oC']==90]

P3_tem25 = X3_tem25['磁芯损耗，w/m3']
P3_tem50 = X3_tem50['磁芯损耗，w/m3']
P3_tem70 = X3_tem70['磁芯损耗，w/m3']
P3_tem90 = X3_tem90['磁芯损耗，w/m3']

P3_tem25.columns = ['25度']
P3_tem50.columns = ['50度']
P3_tem70.columns = ['70度']
P3_tem90.columns = ['90度']

# 材料4
X4 = df4[['温度，oC','励磁波形','磁芯损耗，w/m3']]
X44 = X4[X4['励磁波形']==bx]

X4_tem25 = X44[X44['温度，oC']==25]
X4_tem50 = X44[X44['温度，oC']==50]
X4_tem70 = X44[X44['温度，oC']==70]
X4_tem90 = X44[X44['温度，oC']==90]

P4_tem25 = X4_tem25['磁芯损耗，w/m3']
P4_tem50 = X4_tem50['磁芯损耗，w/m3']
P4_tem70 = X4_tem70['磁芯损耗，w/m3']
P4_tem90 = X4_tem90['磁芯损耗，w/m3']

P4_tem25.columns = ['25度']
P4_tem50.columns = ['50度']
P4_tem70.columns = ['70度']
P4_tem90.columns = ['90度']
```

```python
fig, axs = plt.subplots(1, 4, figsize=(8, 2), sharey=True, dpi=100)

axs[0].boxplot(P1_tem25,positions=[1])
axs[0].boxplot(P2_tem25,positions=[2])
axs[0].boxplot(P3_tem25,positions=[3])
axs[0].boxplot(P4_tem25,positions=[4])
#axs[0].set_title('温度为25')

axs[1].boxplot(P1_tem50,positions=[1])
axs[1].boxplot(P2_tem50,positions=[2])
axs[1].boxplot(P3_tem50,positions=[3])
axs[1].boxplot(P4_tem50,positions=[4])
#axs[1].set_title('温度为50')

axs[2].boxplot(P1_tem70,positions=[1])
axs[2].boxplot(P2_tem70,positions=[2])
axs[2].boxplot(P3_tem70,positions=[3])
axs[2].boxplot(P4_tem70,positions=[4])
#axs[2].set_title('温度为70')

axs[3].boxplot(P1_tem90,positions=[1])
axs[3].boxplot(P2_tem90,positions=[2])
axs[3].boxplot(P3_tem90,positions=[3])
axs[3].boxplot(P4_tem90,positions=[4])
#axs[3].set_title('温度为90')
# 添加标题和坐标轴标签

plt.show()
```
